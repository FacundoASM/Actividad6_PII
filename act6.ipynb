# 1. Calculadora de IMC (Índice de Masa Corporal)

Cree una función que calcule el IMC (Índice de Masa Corporal) de una persona a partir de su peso y altura. La función debe recibir como parámetros el peso (en kg) y la altura (en cm) y devolver el valor del IMC.
***
Opcional: Implemente una segunda función que clasifique el IMC según los estándares de la OMS:

* Bajo peso: IMC < 18.5
* Peso normal: 18.5 <= IMC < 25
* Sobrepeso: 25 <= IMC < 30
* Obesidad: IMC >= 30


```python
def calcular_imc(peso, altura_cm):
    altura_m = altura_cm / 100
    imc = peso / (altura_m ** 2)
    return imc

def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"


peso = float(input('Introduce tu peso en kg: '))
altura = float(input('Introduce tu altura en cm: '))

imc = calcular_imc(peso, altura)
categoria = clasificar_imc(imc)

print(f"IMC: {imc:.2f}")
print(f"Clasificación: {categoria}")
```

# 2. Analizador de números:

Cree una función que reciba un número entero como parámetro y devuelva un diccionario con la siguiente información:

* Si el número es par o impar
* Si el número es positivo, negativo o cero
* Si el número es divisible por 3
* Si el número es divisible por 5
* Si el número es divisible por 3 y por 5
***
Opcional: Implemente la función utilizando una función lambda para cada una de las verificaciones.

```python
def analizar_numero(numero):
    es_par = lambda n: n % 2 == 0
    es_impar = lambda n: n % 2 != 0
    es_positivo = lambda n: n > 0
    es_negativo = lambda n: n < 0
    es_cero = lambda n: n == 0
    divisible_por_3 = lambda n: n % 3 == 0
    divisible_por_5 = lambda n: n % 5 == 0
    divisible_por_3_y_5 = lambda n: n % 3 == 0 and n % 5 == 0

    resultado = {
        "par_o_impar": "par" if es_par(numero) else "impar",
        "positivo_negativo_o_cero": ("positivo" if es_positivo(numero) else "negativo") if not es_cero(numero) else "cero",
        "divisible_por_3": divisible_por_3(numero),
        "divisible_por_5": divisible_por_5(numero),
        "divisible_por_3_y_5": divisible_por_3_y_5(numero)
    }

    return resultado

numero = int(input('Introduce un número entero: '))
resultado = analizar_numero(numero)

print(f"Análisis del número {numero}:")
for clave, valor in resultado.items():
    print(f"{clave.replace('_', ' ').capitalize()}: {valor}")
```

# 3. Calculadora de área y perímetro de figuras geométricas:

Cree funciones para calcular el área y perímetro de las siguientes figuras geométricas:

* Cuadrado
* Triángulo (equilátero, rectángulo e isósceles)
* Círculo
* Rectángulo

Cada función debe recibir las dimensiones necesarias como parámetros y devolver el valor correspondiente del área o perímetro.

```python
import math

# Funciones para el Cuadrado
def area_cuadrado(lado):
    return lado ** 2

def perimetro_cuadrado(lado):
    return 4 * lado

# Funciones para el Triángulo Equilátero
def area_triangulo_equilatero(lado):
    return (math.sqrt(3) / 4) * (lado ** 2)

def perimetro_triangulo_equilatero(lado):
    return 3 * lado

# Funciones para el Triángulo Rectángulo
def area_triangulo_rectangulo(base, altura):
    return (base * altura) / 2

def perimetro_triangulo_rectangulo(base, altura):
    hipotenusa = math.sqrt(base ** 2 + altura ** 2)
    return base + altura + hipotenusa

# Funciones para el Triángulo Isósceles
def area_triangulo_isosceles(base, lados_iguales):
    altura = math.sqrt(lados_iguales ** 2 - (base / 2) ** 2)
    return (base * altura) / 2

def perimetro_triangulo_isosceles(base, lados_iguales):
    return 2 * lados_iguales + base

# Funciones para el Círculo
def area_circulo(radio):
    return math.pi * (radio ** 2)

def perimetro_circulo(radio):
    return 2 * math.pi * radio

# Funciones para el Rectángulo
def area_rectangulo(base, altura):
    return base * altura

def perimetro_rectangulo(base, altura):
    return 2 * (base + altura)

# Función principal
def main():
    while True:
        print("\nSeleccione la figura geométrica:")
        print("1. Cuadrado")
        print("2. Triángulo equilátero")
        print("3. Triángulo rectángulo")
        print("4. Triángulo isósceles")
        print("5. Círculo")
        print("6. Rectángulo")
        print("7. Salir")

        opcion = int(input("Introduce el número de la figura: "))

        if opcion == 1:
            lado = float(input("Introduce el lado del cuadrado: "))
            print(f"Área del cuadrado: {area_cuadrado(lado)}")
            print(f"Perímetro del cuadrado: {perimetro_cuadrado(lado)}")
        
        elif opcion == 2:
            lado = float(input("Introduce el lado del triángulo equilátero: "))
            print(f"Área del triángulo equilátero: {area_triangulo_equilatero(lado)}")
            print(f"Perímetro del triángulo equilátero: {perimetro_triangulo_equilatero(lado)}")
        
        elif opcion == 3:
            base = float(input("Introduce la base del triángulo rectángulo: "))
            altura = float(input("Introduce la altura del triángulo rectángulo: "))
            print(f"Área del triángulo rectángulo: {area_triangulo_rectangulo(base, altura)}")
            print(f"Perímetro del triángulo rectángulo: {perimetro_triangulo_rectangulo(base, altura)}")
        
        elif opcion == 4:
            base = float(input("Introduce la base del triángulo isósceles: "))
            lados_iguales = float(input("Introduce los lados iguales del triángulo isósceles: "))
            print(f"Área del triángulo isósceles: {area_triangulo_isosceles(base, lados_iguales)}")
            print(f"Perímetro del triángulo isósceles: {perimetro_triangulo_isosceles(base, lados_iguales)}")
        
        elif opcion == 5:
            radio = float(input("Introduce el radio del círculo: "))
            print(f"Área del círculo: {area_circulo(radio)}")
            print(f"Perímetro del círculo: {perimetro_circulo(radio)}")
        
        elif opcion == 6:
            base = float(input("Introduce la base del rectángulo: "))
            altura = float(input("Introduce la altura del rectángulo: "))
            print(f"Área del rectángulo: {area_rectangulo(base, altura)}")
            print(f"Perímetro del rectángulo: {perimetro_rectangulo(base, altura)}")
        
        elif opcion == 7:
            print("Saliendo del programa.")
            break
        
        else:
            print("Opción no válida, por favor intente de nuevo.")

if __name__ == "__main__":
    main()
```
